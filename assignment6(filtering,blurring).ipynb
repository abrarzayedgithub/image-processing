1. What is bilateral filtering? How it works, explain?


ans: BIlateral filtering is an advance image processing technique which is used to remove noise and smooth images. The speciality os the bilateral filtering is -when we apply bilateral filtering we get smoother image as much as possible.
It removes the background noise of the image as much as possible.



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/drive/MyDrive/house.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
avg_blur = cv2.blur(image,(5,5))
bilateral = cv2.bilateralFilter(image,5,5,75)
gaussian_blur = cv2.GaussianBlur(image,(5,5),0)
median_blur = cv2.medianBlur(image,5)

plt.figure(figsize=(10,10))
plt.subplot(1,4,1)
plt.title('avg_blur')
plt.imshow(avg_blur,cmap='gray')
plt.axis('off')


plt.subplot(1,4,2)
plt.title('bilateral')
plt.imshow(bilateral,cmap='gray')
plt.axis('off')


plt.subplot(1,4,3)
plt.title('gaussian_blur')
plt.imshow(gaussian_blur,cmap='gray')
plt.axis('off')

plt.subplot(1,4,4)
plt.title('median_blur')
plt.imshow(median_blur,cmap='gray')
plt.axis('off')

plt.show()
```

2. Explain the sigmaColor and sigmaSpace parameter of bilateralFilter() function of OpenCV.

ans: biletaralFilter() is used for edge preserving smoothing. when we apply bilateral filtering,we smooth the image as much as possible while keepng the sharpness of the image.
it contains two important perameters:
1.sigma color
2.sigma space

the sigmacolor works based on the color difference of two pixels.If the color difference of the two pixels is samller than sigmacolor , the two pixels will get mixed with each other.

the sigmaspace works based on the space difference of two pixels. The larger the sigmaSpace value, the wider the blur effect.


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/drive/MyDrive/iiuc mosque.jpg')
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

#apply bilaateral filter
bilateral = cv2.bilateralFilter(image,d=15,sigmaColor=30,sigmaSpace=30)

#convert bilateral filtered image to rgb
bilateral_rgb = cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('original')
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('bilateralfilter')
plt.imshow(bilateral_rgb)
plt.axis('off')

plt.show()



```
